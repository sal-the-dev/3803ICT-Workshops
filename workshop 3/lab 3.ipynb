```python
# Import necessary modules
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression, LogisticRegression

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# suppress warnings
import warnings
warnings.filterwarnings('ignore')

## SECTION 1 - Classification

# Load MNIST digits dataset
digits = datasets.load_digits()
```

```python
# Create feature and target arrays
X = digits.data
y = digits.target

# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# kNN classifier
number_neighbours = []
testing_scores = []
training_scores = []

for i in range(1, 8):
    # create classifier with num neighbours = i
    kNN = KNeighborsClassifier(n_neighbors=i)
    # train classifier on training set (data + target)
    kNN.fit(X_train, y_train)
    number_neighbours.append(i)
    training_scores.append(kNN.score(X_train, y_train))
    testing_scores.append(kNN.score(X_test, y_test))

plt.plot(number_neighbours, testing_scores, label="Testing Accuracy")
plt.plot(number_neighbours, training_scores, label="Training Accuracy")
plt.xlabel("Number of Neighbours")
plt.ylabel("Accuracy")
plt.title("Testing and Training accuracy as a result of number of neighbours")
plt.legend()
```

The disparity between the testing and training accuracy is minimised around k = 2 and k = 3, while maintaining high accuracy scores. After running the sample many times, it appears that k = 2 is the most consistent.

```python
# SVM classifier
kernels = ["linear", "poly", "rbf", "sigmoid"]

for kernel in kernels:
    # create SVM with kernel
    svm = SVC(kernel=kernel)
    # fit SVM to training data
    svm.fit(X_train, y_train)
    # make predictions on test data
    y_pred = svm.predict(X_test)
    print(kernel)
    # show main classification metrics based on actual labels for testing data
    print(classification_report(y_test, y_pred))
    
```

linear, polynomial, and the Radial basis function kernel all appear to have high f-measures, and would be suitable kernels to use for an SVM in this case. However, if one must be chosen, linear appears to be the most consistent (when re-testing on different data splits).

```python
# Naive Bayes
bayes = GaussianNB()
bayes.fit(X_train, y_train)
y_pred = bayes.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
ax = sns.heatmap(conf_matrix, annot=True)
ax.set_xlabel("Predicted label")
ax.set_ylabel("True label")
```

```python
## SECTION 2 - Regression
df = pd.read_csv("gapminder.csv")
df.head()
```

```python
# heatmap visualisation 
sns.heatmap(df.corr(), cmap="RdYlGn")
```

```python
# linear regression
x = df["fertility"]
x2d = x[:, None]
y = df["life"]

regression = LinearRegression()
regression.fit(x2d, y)
print(f'score for fertility: {regression.score(x2d, y)}')

fig, ax = plt.plot(x, y, 'bo', x, regression.predict(x2d), '-k')
plt.xlabel("Fertility")
plt.ylabel("Life Expectency")
plt.show()
```

```python
# linear regression on all
# Region is not included because it is a categorical variable
X = df[["population", "fertility", "HIV", "CO2", "BMI_male", "GDP", "BMI_female", "child_mortality"]]
y = df["life"]

regression_all = LinearRegression()
regression_all.fit(X, y)
print(f'score for all: {regression_all.score(X, y)}')
```

```python
# cross validation score
print(f'cross validation score all: {cross_val_score(regression, x2d, y, cv = 5)}')
print(f'cross validation score fertility: {cross_val_score(regression_all, X, y, cv = 5)}')
```

```python
# load creaitcard csv
df = pd.read_csv("creditcard.csv")
print(df.shape)
df.head()
```

```python
# class and amount comparison
frauds = df.loc[df["Class"] == 1]
non_frauds = df.loc[df["Class"] == 0]
print(f'{len(frauds)} frauds, {len(non_frauds)} non frauds.')

plt.plot(frauds.Amount, frauds.Class, 'o', color="orange", label="Frauds")
plt.plot(non_frauds.Amount, non_frauds.Class, 'bo', label="Non Fraud")
plt.xlabel("Amount")
plt.ylabel("Class")
plt.legend()
```

Fraudulent transactions are almost always of low amounts (between 0 - 5000)

```python
# logistic regression
x = df.drop("Class", axis=1)
y = df.Class
log_regression = LogisticRegression()
log_regression.fit(x, y)
y_pred = log_regression.predict(x)
conf_matrix = confusion_matrix(y, y_pred)

ax = sns.heatmap(conf_matrix, annot=True, fmt='g')
ax.set_xlabel("Predicated label")
ax.set_ylabel("True label")
```
